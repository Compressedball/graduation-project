```python
import numpy as np
import scipy.sparse as sp
```

```python
data_name = "citeseer"

idx_features_labels = np.genfromtxt(f"{data_name}/{data_name}.content", dtype=np.dtype(str))

features = np.array(idx_features_labels[:, 1:-1], dtype=np.float32)

labels = idx_features_labels[:, -1]

idx = np.array(idx_features_labels[:, 0], dtype=np.str_)
```

```python
print("Feature mesage:")
print(type(features))
print(features.shape)
print(features[0])
```

```python
idx_map = {j: i for i, j in enumerate(idx)}

edges_unordered = np.genfromtxt(f"{data_name}/{data_name}.cites", dtype=np.str_)

missing = [
    (src, dst)
    for src, dst in edges_unordered
    if src not in idx_map or dst not in idx_map
]

print("missing edges:", len(missing))
print(missing[:10])   # 看前10个

edges = np.array([
    [idx_map[src], idx_map[dst]]
    for src, dst in edges_unordered
    if src in idx_map and dst in idx_map
])
# coo_matrix((data, (row, col)), shape)
adj = sp.coo_matrix(
    (np.ones(edges.shape[0]), (edges[:, 0], edges[:, 1])),
    shape=(labels.shape[0], labels.shape[0]),
    dtype=np.float32
)
print(type(adj), adj.shape)
```

```python
classes = sorted(set(labels))
class_map = {c: i for i, c in enumerate(classes)}
labels = np.array(list(map(class_map.get, labels)), dtype=np.int32)
print(type(labels))
print(labels.shape)
print(labels[0])
```

```python
sp.save_npz(f"{data_name}/{data_name}_adj.npz", adj)
np.save(f"{data_name}/{data_name}_features.npy", features)
np.save(f"{data_name}/{data_name}_labels.npy", labels)
```
