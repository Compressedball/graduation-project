```python
import scipy.sparse as sp
import numpy as np
import torch
```

```python
data_name = "pubmed"
with open("pubmed/data/Pubmed-Diabetes.NODE.paper.tab") as f:
    header1 = f.readline()
    header2 = f.readline()

print(header1)
print(header2)
feature_names = []

for item in header2.split():
    if item.startswith("numeric:"):
        name = item.split(":")[1]
        feature_names.append(name)

print(len(feature_names))
print(feature_names[-1:])
```

```python
feature_map = {f:i for i, f in enumerate(feature_names)}

features = np.zeros((19717, len(feature_names)), dtype=np.float32)

paper_ids = []
labels = []
with open("pubmed/data/Pubmed-Diabetes.NODE.paper.tab") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) < 3:
            print("Invalid line:", parts)
            continue
        if not parts[0].isdigit():
            print("Invalid paper ID:", parts)
            continue
        paper_id = parts[0]
        label = int(parts[1].split("=")[1])
        labels.append(label)
        paper_ids.append(paper_id)
```

```python
labels = labels - np.min(labels)  # Shift labels to start from 0
print(f"Total papers: {len(paper_ids)}, Total labels: {len(labels)}")
print(labels.min(), labels.max())

idx_map = {j:i for i,j in enumerate(paper_ids)}
print(f"Index map size: {len(idx_map)}")

with open("pubmed/data/Pubmed-Diabetes.NODE.paper.tab") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) < 3:
            print("Invalid line:", parts)
            continue
        if not parts[0].isdigit():
            print("Invalid paper ID:", parts)
            continue
        for item in parts[2:]:
            if item.startswith("w-"):
                name, value = item.split("=")
                if name in feature_map:
                    features[idx_map[parts[0]], feature_map[name]] = float(value)

print(f"Features shape: {features.shape}")
print(features[-5:])
```

```python
edges = []
with open("pubmed/data/Pubmed-Diabetes.DIRECTED.cites.tab") as f:
    for line in f:
        parts = line.strip().split()
        if not parts[0].isdigit():
            print("Invalid paper ID:", parts)
            continue
        paper1 = parts[1].split(":")[1]
        paper2 = parts[3].split(":")[1]
        if paper1 in idx_map and paper2 in idx_map:
            edges.append((idx_map[paper1], idx_map[paper2]))
        else:
            print(f"Skipping edge with missing paper: {paper1} or {paper2}")

edges = np.array(edges)
print(f"Total edges: {edges.shape}")
# coo_matrix((data, (row, col)), shape)
adj = sp.coo_matrix(
    (np.ones(edges.shape[0]), (edges[:, 0], edges[:, 1])),
    shape=(labels.shape[0], labels.shape[0]),
    dtype=np.float32
)
print(type(adj), adj.shape)
```

```python
sp.save_npz(f"{data_name}/{data_name}_adj.npz", adj)
np.save(f"{data_name}/{data_name}_features.npy", features)
np.save(f"{data_name}/{data_name}_labels.npy", labels)
```
